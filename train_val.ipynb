```python
data_np = data_df.to_numpy()
print('shape:',data_np.shape)
x_np = data_np[:,:-1]
y_np = data_np[:,-1:]
le = preprocessing.LabelEncoder()
le.fit(y_np)
y_np_num = le.transform(y_np) # convert labels to numbers 
#y_np_inverse = le.inverse_transform(y_np_num) # check if converting from numbers to labels give correct results 
print('x shape: %s, y shape: %s' % (x_np.shape,y_np.shape))
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(x_np,y_np,test_size = 0.2,random_state=42, shuffle = True)
print('train shape: %s, test shape: %s' % (X_train.shape,X_test.shape))
print('y_train shape:',y_train.shape)
```
